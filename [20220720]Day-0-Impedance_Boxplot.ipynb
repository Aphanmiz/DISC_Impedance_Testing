```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
from statsmodels.formula.api import ols
import statsmodels.api as sm
import statsmodels

def significance_plot(x1, x2, y, p_value):
    h=0
    plt.plot([x1, x1, x2, x2], [y, y + h, y + h, y], lw=1.5, color='black')
    # h=h-1
    if p_value <= 0.001:
        plt.text((x1 + x2) * .5, y + h, "***", ha='center', va='bottom', fontsize=11)
    elif p_value <= 0.01:
        plt.text((x1 + x2) * .5, y + h, "**", ha='center', va='bottom', fontsize=11)
    else:
        plt.text((x1 + x2) * .5, y + h, "*", ha='center', va='bottom', fontsize=11)

data=pd.read_csv('Day0_imped_mutual.csv')

# fig=plt.figure(figsize=[15, 6])

ax=sns.boxplot(x = "type", y = "impedance", data=data, color='lightgrey',showmeans=True,    meanprops={"marker":"_",
                       "markerfacecolor":"black",
                       "markeredgecolor":"black",
                      "markersize":"10"})#meanprops={'color': 'red', 'ls':'.', 'lw': 5})
ax=sns.swarmplot(x = "type", y = "impedance", data=data, size=10, dodge=True, linewidth=1, edgecolor="gray", alpha=.8)

ax.set_ylabel('Day', fontsize=18, fontname='Arial')
ax.set_xlabel('\n Gap Size', fontsize=18, fontname='Arial')

medians = data.groupby(['type'])['impedance'].median()
vertical_offset = data['impedance'].median() * 0.05 # offset from median for display

for xtick in ax.get_xticks():
# plot the median
#     ax.text(xtick,medians[xtick] + vertical_offset,medians[xtick], 
#             horizontalalignment='center',size='small',color='black',weight='semibold')
    print("median:", medians[xtick])


plt.xticks(fontsize=15, fontname='Arial')

plt.subplots_adjust(wspace=0, hspace=0)
plt.show()
```
